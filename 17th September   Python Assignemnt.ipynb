```python
# Get the two numbers from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Initialize the GCD variable
gcd = 1

# Use a for loop to iterate from 1 to the smaller of the two numbers
for i in range(1, min(num1, num2) + 1):
    # Check if the current number divides both numbers evenly
    if num1 % i == 0 and num2 % i == 0:
        # Update the GCD
        gcd = i

# Print the GCD
print(f"The GCD of {num1} and {num2} is: {gcd}")
```

21. Explain the purpose of the "continue" and "break" statements within if-else loops.


The continue and break statements are used in Python loops (like for and while loops) to alter the flow of control in a loop. They can be used within if-else blocks inside these loops.

Here’s what they do:

**continue:** When the continue statement is encountered, the current iteration of the loop is terminated immediately, and execution moves to the next iteration of the loop. Any code following the continue statement within the loop is skipped for the current iteration.  
**break:** The break statement, when encountered, causes an immediate exit from the loop, regardless of the loop’s condition. Execution continues at the first statement after the loop. Any code following the break statement within the loop, or any remaining iterations, are skipped.

```python
for i in range(1, 10):
    if i == 3:
        continue
    elif i == 7:
        break
    print(i)
```

1. Explain the purpose of the `map()` function in Python and provide an example of how it can be used to apply a function to each element of an iterable.

The map() function in Python is used to apply a function to each item in an iterable (like a list or a tuple) and returns a map object (which is an iterable itself). You can convert this map object to a list or another iterable type if needed.

Here’s a simple example of how map() can be used:

```python
def square(n):
    return n ** 2

numbers = [1, 2, 3, 4, 5]
squared = map(square, numbers)

print(list(squared))  # Output: [1, 4, 9, 16, 25]
```

3. How does the `map()` function differ from a list comprehension in Python, and when would you choose one over the other?

The map() function and list comprehension in Python both provide a way to apply a function to each item in an iterable, but they do so in slightly different ways:   

**map() function:** This is a built-in Python function that takes in two or more arguments: a function and one or more iterables. map() applies the function to each item of the iterable and returns a map object. You can convert this map object to a list or another iterable type if needed.    
**List comprehension:** This is a concise way to create lists based on existing lists (or other iterables). In list comprehension, you can include an expression followed by a for statement inside square brackets []. You can also include if conditions.   

As for when to use one over the other, it depends on your specific needs and personal preference:

**Readability:** List comprehensions are often easier to read and understand than map() especially for simple operations. However, for more complex operations, a well-named function with map() might be more readable.  
**Performance:** For large datasets, map() can be more efficient as it returns an iterator and doesn’t generate the whole list in memory at once.   
**Flexibility:** List comprehensions can be more flexible as they can handle more complex expressions and conditions.



Here’s an example that demonstrates the difference:

```python
# Using map()
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x**2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16, 25]

# Using list comprehension
numbers = [1, 2, 3, 4, 5]
squared = [x**2 for x in numbers]
print(squared)  # Output: [1, 4, 9, 16, 25]
```

# Recursion:-

1. Explain the concept of recursion in Python. How does it differ from iteration?

Recursion in Python is a process in which a function calls itself as a subroutine. This allows the function to be repeated several times, as it can call itself during its execution.

Here’s an example of a recursive function in Python:

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)
```

**Key Differences Between Recursion and Iteration:**  

**Concept:** Recursion is a method where the solution to a problem depends on solutions to smaller instances of the same problem. Iteration is a general term for taking each item of something, one after another.  
**Functionality:** In recursion, a function calls itself until the base case is reached. In iteration, a set of instructions is repeated a certain number of times.   
**Efficiency:** Recursion can be slower and requires more memory because it needs to manage the function call stack.Iteration, being a simple loop, tends to be faster and more memory-efficient.   
**Ease of Use:** Recursion can make code look cleaner and easier to understand. Iteration can be more straightforward to grasp, especially for simple repetitions.   
**Risk of Errors:** Recursion has the risk of causing a stack overflow if the base case isn’t met in recursive calls. Iteration doesn’t have this risk, but infinite loops can occur if the loop condition is always true.

9. Write a recursive Python function to find all permutations of a given string.

```python
def permutations(string):
    if len(string) == 0:
        return [""]
    else:
        result = []
        for i in range(len(string)):
            char = string[i]
            remaining_chars = string[:i] + string[i+1:]
            perms = permutations(remaining_chars)
            for perm in perms:
                result.append(char + perm)
        return result

string = input("Enter a string: ")
permutations_list = permutations(string)
print("All permutations of the string are:", permutations_list)
```

```python
def combinations(lst):
  if len(lst) == 0:
    return [[]]
  else:
    result = []
    for i in range(len(lst)):
      element = lst[i]
      remaining_elements = lst[:i] + lst[i+1:]
      combs = combinations(remaining_elements)
      for comb in combs:
        result.append([element] + comb)
    return result

lst = [1, 2, 3, 4]
combinations_list = combinations(lst)
print("All possible combinations of the list are:", combinations_list)
```

3. Explain the concept of higher-order functions in Python, and provide an example.

**What are Higher-Order Functions?:**  
A higher-order function is a function that meets one or more of the following criteria:  
  * Allows other functions to be passed to it as parameters.
  * Can return other functions as a result.
  *Creates new functions within its body.
  
These functions are powerful tools for abstraction, code reuse, and functional programming.

**Examples of Higher-Order Functions:**  
Here are some scenarios where higher-order functions are commonly used:
1. Functions as Arguments:
You can pass functions as arguments to other functions.  

Example:

```python
def greet(func):
    greeting = func("Hi, I am created by a function passed as an argument.")
    print(greeting)

def shout(text):
    return text.upper()

def whisper(text):
    return text.lower()

greet(shout)  # Output: "HI, I AM CREATED BY A FUNCTION PASSED AS AN ARGUMENT."
greet(whisper)  # Output: "hi, i am created by a function passed as an argument."
```

2. Returning Functions:
Functions can also return other functions.

Example:

```python
def create_adder(x):
    def adder(y):
        return x + y
    return adder

add_15 = create_adder(15)
print(add_15(10))  # Output: 25
```

3. Decorators:
Decorators are a common use of higher-order functions.
They allow you to modify the behavior of a function without permanently changing it.

Example:

```python
def my_decorator(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
# Output:
# Before function execution
# Hello!
# After function execution
```

4. Create a Python function that takes a list of numbers and a function as arguments, applying the function to each element in the list.

```python
def apply_function_to_list(numbers, func):
  """
  Applies a function to each element in a list.

  Args:
    numbers: A list of numbers.
    func: A function that takes a number as input and returns a number.

  Returns:
    A list containing the results of applying the function to each element in the list.
  """
  results = []
  for number in numbers:
    results.append(func(number))
  return results

# Example usage
def square(x):
  return x ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = apply_function_to_list(numbers, square)
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

3. How does the `filter()` function work in Python, and when would you use it?

1. **What is filter()?:**  
The filter() function is a built-in Python function that allows you to process an iterable (such as a list, tuple, or set) and extract items that satisfy a given condition.
It performs a filtering operation by applying a specified function to each element of the iterable.
The result is a new iterable containing only the items that meet the condition.

2. **Syntax:**
The syntax of the filter() function is as follows:

  filter(function, iterable)  
Parameters:  
function: A function that tests each element of the iterable.
iterable: The sequence (e.g., list, tuple) to be filtered.

3. **How filter() Works:**
The filter() function applies the specified function to each item in the iterable.
If the function returns True for an item, that item is included in the result.
If the function returns False, the item is excluded.
The result is an iterator containing the filtered items.  

4. Example:
Let’s say we have a list of ages, and we want to filter out only the ages that are 18 or above:

```python
ages = [5, 12, 17, 18, 24, 32]

def is_adult(age):
    return age >= 18

adults = filter(is_adult, ages)
adult_list = list(adults)

print("Adults:", adult_list)
# Output: Adults: [18, 24, 32]
```

5. **Use Cases:**  
You can use filter() when:  
You need to extract specific elements from a collection based on a condition.
You want to create a new iterable containing only the items that meet certain criteria.

2. Describe some best practices for naming functions and variables in Python, including naming conventions and guidelines.

1. **Use Descriptive Names:**   
Choose meaningful names that reflect the purpose of the variable or function.
Avoid generic names like temp, data, or single-letter variables (unless they have a specific context).  

2. **Follow Naming Conventions:**  
Python has widely accepted naming conventions:  
Snake Case: Use lowercase letters and separate words with underscores (_). Example: user_name, total_count.  
Camel Case: Start with lowercase and capitalize each subsequent word. Example: firstName, totalCount.   
Pascal Case: Start with an uppercase letter and capitalize each subsequent word. Used for class names. Example: Person, CarModel.  

3. **Function Names:**  
Use lowercase letters and separate words with underscores.  
Be descriptive about what the function does.  
Example: calculate_total_price(), validate_user_input().  

4. **Variable Names:**  
Follow the same conventions as function names.  
Use descriptive names for variables.  
Example: customer_name, item_quantity.  

5. **Constants:**  
Use uppercase letters and separate words with underscores.  
Constants should remain unchanged throughout the program.  
Example: MAX_RETRY_ATTEMPTS, PI.  

6. **Avoid Ambiguity:**  
Be clear about the purpose of the variable or function.  
Avoid using vague names like result, value, or data.

7. **Avoid Reserved Words:**  
Do not use Python’s reserved keywords (e.g., if, else, while, for) as variable or function names.  

8. **Be Consistent:**   
Maintain consistent naming conventions throughout your codebase.   
If you’re working on an existing project, follow the existing style.  

9. **Comments and Docstrings:**   
Include docstrings (multi-line comments) to explain the purpose and usage of functions and classes.  
Use inline comments to clarify complex logic or edge cases.  

10. **Avoid Acronyms and Abbreviations:**  
Favor descriptive names over abbreviations.   
Exceptions: Common acronyms like HTTP, URL, or XML.   

11. **Avoid Single-Character Variables:**  
Unless they represent a well-known concept (e.g., i for loop index), avoid single-letter variables.

12. **Avoid Magic Numbers:**
Assign magic numbers (constants) to named variables.  
Example: Instead of total_price = 100 * quantity, use unit_price = 100 and total_price = unit_price * quantity.
